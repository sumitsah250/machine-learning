Logistic Regression importing libraries

```python
import numpy as np
```

```python
class Logistic_Regression():
    #declaring  learning rate and number of iteration(Hyperpapameters)
    def __init__(self,learning_rate,no_of_iteration):
        self.learning_rate=learning_rate
        self.no_of_iteration=no_of_iteration
        #fit function to train the model to the dataset
    def fit(self,X,Y):
        #number of datta pointts in the dataset ( number of rows)---- m
        #number of input feature in the dataset (number of columns)--> n
         self.m,self.n=X.shape
         
         #initiating the weight and bis values
         self.w=np.zeros(self.n)
         self.b=0
         self.X=X
         self.Y=Y 
         
         #implementing gradient Descent for optimization
         for i in range(self.no_of_iteration):
            self.update_weights()
                     
    def update_weights(self):
        #Y_hat formuls(Sigmoid function
    
        Y_hat = 1/(1+np.exp(-(self.X.dot(self.w)+self.b)))  # wX+b
        #derivatives
        
        dw=(1/self.m)*np.dot(self.X.T,(Y_hat-self.Y))
        db=(1/self.m)*np.sum((Y_hat-self.Y))
        
        #updating the weights and bias using gradient descent
        self.w = self.w - self.learning_rate*dw
        self.b= self.b - self.learning_rate*db
        
    #writing the sigmoid equation and decision boundary
    
    def predict(self,X):
        Y_pred =1/(1+np.exp(-(self.X.dot(self.w)+self.b))) 
        Y_pred= np.where(Y_pred>0.5,1,0)
        return Y_pred
        
    
```

```python
features = diabetes_dataset.drop(columns='Outcome',axis=1)
target= diabetes_dataset['Outcome']
print(target)
```

```python
scaler = StandardScaler()
scaler.fit(features)
```

```python
standrdized_data=scaler.transform(features)
print(standrdized_data)
```

```python
features=standrdized_data
target=diabetes_dataset['Outcome']

```

```python
X_train,X_test,Y_train,Y_test= train_test_split(features,target,test_size=0.2,random_state=2)
```

```python
classfier = Logistic_Regression(learning_rate=0.01,no_of_iteration=1000)
```

```python
classfier.fit(X_train,Y_train)
```

```python
X_train_prediction=classfier.predict(X_train)
training_data_accuracy = accuracy_score(Y_train,X_train_prediction)
```

```python
print("the Accuracy score of training data is :",training_data_accuracy)
```

```python
X_train_prediction=classfier.predict(X_test)
X_test_data_accuracy = accuracy_score(Y_train,X_train_prediction)
```
